Для дальнейшей работы отберем только пустые значения в колонке с датами доставки и удалим дубликаты по колонке order_id

Посмотрим на нулевые значения колонок в таблице с отсутствующими датами доставки заказов:

Далее необходимо исключить неоплаченные заказы, т.к. причиной недоставки 146 заказов является их неоплата 
перед этим необходимо увидеть, что 14 из 160 необплаченных заказов все таки были доставлены на каких-то особых условиях 

```python
olist_orders_dataset.query('order_approved_at=="NaN" and order_delivered_customer_date!="NaN"')['order_purchase_timestamp'].count()
```

Исключим доставленные заказы, оставшиеся сгруппируем по указанному статусу и посчитаем количество 
недоставленных заказов

```python
all_reasons = items_orders.groupby('order_status', as_index=False)\
                          .agg({'order_id': 'count'}).rename(columns={'order_id': 'all'})

all_reasons
```

Посмотрим на недоставленные заказы, у которых дата передачи заказа партнеру по логистике превышает 
время передачи заказа в логистическую службу.

Несмотря на то, что в данных указан различный статус для 1045 заказов, мы видим, что причиной их недоставки
было отсутствие товаров на складе на момент отгрузки, т.е товары не поступили от поставщика в доставку  для передачи клиенту

```python
shipping_reasons = items_orders.query('shipping_limit_date>order_delivered_carrier_date')\
                               .groupby('order_status', as_index=False)['order_id'].count()\
                               .rename(columns={'order_id': 'shipping'})

shipping_reasons
```

Теперь посмотрим на недоставленные заказы, по которым обещанная дата доставки оказалась раньше, 
чем дата передачи заказа в логистическую службу.

Параллельно выберем заказы, по которым отсутствует дата передачи заказа в логистику и посчитаем те, у которых 
дата передачи заказа партнеру по логистике превышает обещанную дату доставки.

Всё это вместе будем считать просроченными заказами.

Таким образом у нас есть 21 просроченный заказ. 

Скорее всего, по ним тоже товаров не оказалось на складе на момент отгрузки, либо ошибочно была назначена более ранняя дата доставки клиенту, чем это возможно было сделать.

Клиент мог отказаться от этих товаров изза поздней доставки и поэтому заказы не были доставлены

```python
overdue_reason = items_orders.query('order_estimated_delivery_date<order_delivered_carrier_date\
 or (order_delivered_carrier_date=="NaN" and shipping_limit_date>order_estimated_delivery_date)')\
                             .groupby('order_status', as_index=False)['order_id'].count()\
                             .rename(columns={'order_id': 'overdue'})

overdue_reason
```

Из 1008 заказов, по которым отсутствует дата передачи заказа в логистику мы исключили те, у которых 
дата передачи заказа партнеру по логистике превышает обещанную дату доставки
посмотрим на остальнае

Получается, что в данном случае товары для заказов не отмечены на складе для дальнейшей доставки клиенту
таких заказов 995. 

Они оплачены, поставщик успевал их привезти в срок так, чтобы компания потом успела 
отправить их клиенту, но они не были отправлены, предположительно, это могут оказаться товары плохого качества

```python
damage_items = items_orders.query('order_delivered_carrier_date=="NaN"\
                    and shipping_limit_date<order_estimated_delivery_date')\
                           .groupby('order_status', as_index=False)['order_id'].count()\
                           .rename(columns={'order_id': 'damage'})

damage_items
```

Посмотрим, сколько осталось недоставленных заказов по другим причинам

```python
other_reasons = items_orders.query('shipping_limit_date<=order_delivered_carrier_date\
                   and order_estimated_delivery_date>=order_delivered_carrier_date')\
                            .groupby('order_status', as_index=False)['order_id'].nunique()\
                            .rename(columns={'order_id': 'other_reasons'})

other_reasons
```

Объединим все причины недоставленных заказов в одну таблицу с общей таблицей недоставленных заказов

```python
all_reasons = all_reasons.merge(shipping_reasons, on='order_status', how='left')\
                         .merge(overdue_reason, on='order_status', how='left')\
                         .merge(damage_items, on='order_status', how='left')\
                         .merge(other_reasons, on='order_status', how='left')\
                         .fillna('')

all_reasons
```

Необходимо написать функцию, которая вернет название причины недоставки каждому заказу по условию.

Будем сравнивать даты, которые проверили ранее

```python
def reasons(shipping_limit_date, order_delivered_carrier_date, order_estimated_delivery_date):

    if shipping_limit_date>order_delivered_carrier_date:
        return 'shipping_reasons'   

    elif  (order_delivered_carrier_date!="NaT" and order_estimated_delivery_date<order_delivered_carrier_date):
        return 'overdue_reason' 
    elif (shipping_limit_date>=order_estimated_delivery_date):
        return 'overdue_reason' 
 
    elif shipping_limit_date<order_estimated_delivery_date:
        return 'damage_items'      

    else: 
        return 'other_reasons'
```

Выведем в отдельную колонку причины недоставки каждого заказа, которые смогли определить

```python
items_orders['Reasons'] = items_orders.apply(lambda x: reasons(x.shipping_limit_date, x.order_delivered_carrier_date, x.order_estimated_delivery_date), axis=1)
items_orders.tail(1)
```

Теперь выведем информацию, сколько товаров в месяц в среднем не доставляется

Группируем в разбивке по причинам помесячно

```python
reasons_by_month = items_orders.groupby(['Reasons', 'Period'], as_index=False)['order_id'].count()\
                               .pivot(index='Period', columns='Reasons', values='order_id')
reasons_by_month
```

Считаем среднее ежемесяное количество недоставленных покупателю заказов

```python
reasons_by_month.fillna(0).mean()
```

# 3. По каждому товару определить, в какой день недели товар чаще всего покупается

Для дальнейшей работы нам понадобится копия таблицы товаров, соединенная с таблицей заказов

```python
orders_items = olist_order_items_dataset.merge(olist_orders_dataset, on='order_id')
orders_items.tail(1)
```

Введем ограничение по признаку "Покупка"

```python
orders_items = orders_items.query('order_delivered_customer_date!="NaN"')
orders_items.tail(1)
```

Для выполнения задания нам понадобится колонка с днями недели по каждому заказу
в этом задании нам интересно активность пользователей по дням недели, её мы будем проверять по колонке 
"время создания заказа", потому что именно в этот день клиент захотел что-то приобрести и совершил заказ

Для каждого уникального товара посчитаем, сколько раз клиенты его заказали в каждый из дней недели

Дополнительно можем видеть на графике, что активность клиентов по дням недели примерно одинакова,
но в понедельник и вторник число заказов немного больше, чем в другие дни
детализация по максимальному количеству заказов для каждого отдельного товара по дням недели - в таблице max_orders

```python
ax = sns.barplot(x = "day_of_week", y = "number_items", data=max_orders)
ax.set_xticklabels(max_orders.day_of_week, rotation=45)
sns.despine()
```

# 4. Сколько у каждого из пользователей в среднем покупок в неделю (по месяцам)? 
Не стоит забывать, что внутри месяца может быть не целое количество недель. 
Например, в ноябре 2021 года 4,28 недели. И внутри метрики это нужно учесть

Снова будем работать с таблицей пользователей, соединенной с таблицей заказов.

Нам понадобится копия таблицы customers_orders, в которой мы планиируем избавиться от пустых значений в 
колонке order_delivered_customer_date, выбирая только те строки датафрейма, где есть дата доставки.

Будем делать анализ поведения пользователей на базе информации о датах доставки, потому что ранее мы выяснили,
что доставленные товары оплачены клиентами (кроме 14 случаев) и получены, т.е. покупка состоялась


```python
customers_orders1 = customers_orders.query('order_delivered_customer_date!="NaN"')

customers_orders1.tail(1)
```

Добавляем колонку времени доставки заказа (говоря о покупке, ориентируемся на эту колонку) в формате "месяц-год" (анализируем поведение клиентов в разрезе месяцев, поэтому точная дата до дня нам не нужна) и преобразуем эту колонку в формат datetime.

Так же, нам нужна информация с количеством недель в каждом месяце, её мы получим путем деления количества 
дней в каждом месяце на 7. 

И добавим колонку с количеством недель в датафрейм

```python
customers_orders1["Period"] = pd.PeriodIndex(customers_orders1.order_delivered_customer_date, freq='M')
customers_orders1["numbers_weeks"] = (customers_orders1["order_delivered_customer_date"].apply(lambda x: pd.Period(x, freq='S').days_in_month)/7).round(2)

customers_orders1.tail(1)
```

Теперь можно посчитать число покупок, сделанных каждым уникальным клиентом в разрезе месяцев 
группировка по количеству недель в месяце нужна для будущих расчетов


```python
purchases_weekly = customers_orders1.groupby(['customer_unique_id', 'Period', 'numbers_weeks'], as_index=False)\
                                    .agg({'order_id': 'count'})\
                                    .rename(columns={'order_id': 'quantity_purchases'})
purchases_weekly.tail(1)
```

Теперь можно посчитать, сколько каждый пользователь в среднем совершил покупок в неделю (по месяцам)? 

```python
purchases_weekly = purchases_weekly.assign(average_purchases_week = lambda x: purchases_weekly.quantity_purchases/purchases_weekly.numbers_weeks).round(2)

purchases_weekly[['customer_unique_id', 'Period', 'average_purchases_week']].tail()
```

# 5.1. Выполните когортный анализ пользователей

Опять будем работать с таблицей пользователей, соединенной с таблицей заказов.

Нам понадобится копия таблицы customers_orders1

Оставляем только нужные колонки для дальнейшей работы, нам нужна информация об уникальном номере клиента, 
уникальном номере заказа и дате доставки в формате "месяц-год"

```python
customers_orders2 = customers_orders1[['customer_unique_id', 'order_id', 'Period']]

customers_orders2.tail(1)
```

Определяем дату, когда произошла первая покупка клиента (временем первой покупки клиента считаем минимальную 
дату для каждого уникального клиента из колонки Period), а колонку с месяцем первой покупки назовем Start 
и это будет когортные месяцы, с помощью которых мы увидим, когда мы приобрели каждого первого клиента

```python
Start = customers_orders2.groupby('customer_unique_id')['Period'].min().rename('Start')
```

Объединяем наш датафрейм и информацию про время первой покупки

```python
customers_orders2 = pd.merge(customers_orders2, Start, left_on='customer_unique_id', right_index=True)

customers_orders2.tail(1)
```

Теперь нам нужно посмотреть на разницу в когортах, сколько месяцев клиенты были активны.

Напишем функцию для получения отдельных частей (года и месяца) для колонок с датами любой покупки клиента 
и датами первой покупки клиента. 

Нам понадобится узнать количество месяцев, прошедших с момента первой покупки 
клиента, т.к. мы хотим видеть информацию помесячно. 


```python
def get_date_part(df, column):
    month = df[column].dt.month
    year = df[column].dt.year
    return month, year
```

```python
Period_month,Period_year = get_date_part(customers_orders2, 'Period')
Start_month,Start_year = get_date_part(customers_orders2, 'Start')
```

Создадим колонку (Time_diff), в которой покажем разницу между датой очередной покупки клиента и датой его   
первой покупки (на основании элементов, которые мы выделили ранее), 
из которой становится понятно, спустя какой срок клиент возвращается для повторной покупки.

Это будет индекс когорты

```python
year_diff = Period_year - Start_year
month_diff = Period_month  - Start_month
```

Чтобы получить информацию в месяцах, умножаем разницу лет на 12.

Для некоторых клиентов месяц покупки совпадает с месяцем первой покупки, и это не нулевой период, 
поэтому к разнице месяцев прибавим 1


```python
customers_orders2['Time_diff'] = year_diff*12+month_diff+1

customers_orders2.head()
```

Мы рассматриваем ежемесячные когорты, поэтому сгруппируем клиентов по дате первой покупки, 
вторая группировка по периоду активности клиента нужна для того, чтобы мы понимали, как долго каждый 
уникальный клиент был активен с момента первой покупки, и посчитаем количество активных клиентов

```python
cohorts = customers_orders2.groupby(['Start', 'Time_diff'], as_index=False)['customer_unique_id'].agg(['nunique'])\
                           .rename(columns={'nunique': 'Active_cust'})\
                           .reset_index()
cohorts.tail()
```

Создадим когортную таблицу, в которой строками будут месяцы первой покупки, т.е. мы увидим активность клиентов,
столбцы - это удержание клиентов с момента первого приобретения, т.е. в каждый момент времени мы видим, 
сколько у нас активных клиентов спустя определенный промежуток времени после первой покупки

```python
cohort_table = cohorts.pivot_table(index = 'Start', columns = 'Time_diff', values = 'Active_cust')
cohort_table
```

Визуализируем данные когортной таблицы с помощью тепловой карты

```python
plt.figure(figsize=(21, 10))
sns.heatmap(cohort_table, annot=True, vmin=0, vmax=0.01)
```

Мы видим, что активнее всего ведут себя клиенты в первые месяцы покупки, далее активность снижается

Например, мы хотим понять, сколько заказов клиенты совершают в течение года после первой покупки.

Отфильтруем колонку Time_diff в датафрейме customers_orders2, оставив только значения, не превышающие 12 месяцев

```python
one_year = customers_orders2.query('Time_diff<=12')
one_year.tail(1)
```

Теперь сгруппируем заказы по дате первой покупки и посчитаем количество заказов 

```python
one_year_orders = one_year.groupby(['Start', 'Time_diff'])['order_id'].agg(['count'])\
                          .rename(columns={'count': 'Quantity_orders'})\
                          .reset_index()
one_year_orders.tail()
```

Теперь сформируем когортную таблицу и визуализируем данные по количеству заказов,

 Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию.

R - время от последней покупки пользователя до текущей даты, 
F - суммарное количество покупок у пользователя за всё время, 
M - сумма покупок за всё время. 

Подробно опиши, как ты создавал кластеры. 
Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров

Пример такого описания: RFM-сегмент 132 (recency=1, frequency=3, monetary=2) 
имеет границы метрик recency от 130 до 500 дней, frequency от 2 до 5 заказов в неделю, 
monetary от 1780 до 3560 рублей в неделю.

Объединим три датафрейма и оставим нужные столбцы

```python
customers_orders_items_ = (olist_customers_datase.merge(olist_orders_dataset, on='customer_id')\
                                                 .merge(olist_order_items_dataset, on='order_id'))\
                                                 .query('order_purchase_timestamp!="NaN"')

customers_orders_items_.tail(1)
```

Нам понадобится информация об уникальном id клиента, дате создания заказа, номере заказа и цене

Посмотрим на графики распределения числа уникальных клиентов и выручки по месяцам за все время,

Будем считать информацию из колонки "price" выручкой, т.к. нам не известна информация

Сократим таблицу и будем анализировать данные в течение выбранного периода [09.2017 - 08.2018]

```python
customers_orders_items = customers_orders_items.query('Period>="2017-09" and Period<="2018-08"')
customers_orders_items.tail(1)
```

Установим текущую дату на один день после последней покупки, считаем, что это текущий момент времени. 

Эта дата будет использоваться в качестве ориентира для расчета показателя новизны.

```python
NOW = customers_orders_items['order_purchase_timestamp'].max() + timedelta(days=1)

NOW
```

Проверим период продолжительностью в пределах имеющихся дат

```python
period = (customers_orders_items['order_purchase_timestamp'].max()) - (customers_orders_items['order_purchase_timestamp'].min())

period
```

Добавим столбец с количеством дней с даты последней покупки по текущий момент.

Найдем минимум этого столбца для каждого клиента для расчета времени с момента его последней покупки

```python
customers_orders_items['days'] = customers_orders_items['order_purchase_timestamp'].apply(lambda x: (NOW - x).days)

customers_orders_items.tail(1)
```

Сформируем таблицу RFM, в которой будут данные о времени с момента последней покупки, о количестве заказов и о сумме заказов за исследуемый период по каждому клиенту.

Для клиентов, которые покупали несколько раз берем минимальное время из имеющихся данных.

В качестве источника информации для частоты покупок используем число уникальных идентификаторов заказов (номер чека), в качестве источника информации для суммы покупок используем суммы из колонки "price". 

Заметим, что а данных не хватает количества купленных товаров, чтобы можно было посчитать выручку по каждому клиенту

```python
RFM = customers_orders_items.groupby(['customer_unique_id'])\
                            .agg({'days': 'min', 'order_id': 'nunique', 'price': 'sum'})\
                            .rename(columns={'days': 'Recency', 'order_id': 'Frequency', 'price': 'Monetary'})\
                            .reset_index()
RFM.tail()
```

Будем создавать границы метрики 'Monetary' вручную экспертно на основании визуальных наблюдений: 

(2200, 13440] - ранг 3 - в основном эти клиенты закупались в одном месяце, оптом (суммы крупные) 
                          и многие относительно недавно (1 самый крупный клиент купил около года назад)
                          
(200 : 2200] - ранг 2 - увеличивается частота покупок клиентов, потраченные суммы выше среднего

(0 : 200] - ранг 1 - основная масса - клиенты, потратившие маленькие суммы а разный период с разной частотой

Поскольку метрика Recency распределена равномерно по всем данным, можно разделить ее на любое количество интервалов пропорционально количеству наблюдений.

Для Recency лучше, когда значения более низкие, т.е., когда от момента последней покупки прошло меньше дней.

Решено разделить на 3 интервала:

(0.999, 110.0] - ранг 3 - клиенты, совершившие покупку недавно

(110.0, 219.0] - ранг 2 - относительно недавние клиенты

(219.0, 365.0] - ранг 1 - прошлогодние клиенты, покупали давно

```python
RFM['R'] = pd.qcut(RFM['Recency'], q=3, labels=[3, 2, 1]).astype(object)

RFM.groupby('R').agg({'customer_unique_id': 'count'})
```

Для Frequency создадим категории отдельно на основании границ метрики, полученных вручную.

Чем больше частота покупок, тем лучше

```python
RFM['F'] = pd.cut(RFM['Frequency'], bins=[0, 2, 5, 11], labels=[1, 2, 3]).astype(str)

RFM.groupby('F').agg({'customer_unique_id': 'count'})
```

Создадим категории для метрики Monetary на основании границ, рассчитанных вручную.

Чем больше потраченная сумма, тем лучше

```python
RFM['M'] = pd.cut(RFM['Monetary'], bins=[0, 250, 1500, 13440], labels=[1, 2, 3]).astype(str)

RFM.groupby('M').agg({'customer_unique_id': 'count'})
```

Объединяем метрики R, F и M в оценку RFM.

```python
RFM['RFM_Score'] = RFM['R'].map(str)+ RFM['F'].map(str) + RFM['M'].map(str)
RFM.tail()
```

Для себя посмотрим на % распределения клиентов по сегментам

```python
a = RFM.groupby('RFM_Score').agg({'customer_unique_id': 'count', 'Monetary': 'sum'}).reset_index()
```

```python
a['persentage'] = ((a.Monetary / a.Monetary.sum()) * 100).round(2)
a
```

# Кластеризация

Описание сегментов: 

VIP - Купил недавно, часто покупает и тратит больше всех

Loyal - Недавние покупатели, либо покупатели на регулярной основе

Potential loyal - Нуждающиеся во Внимании - Выше среднего по частоте покупок и денежными значениями, либо недавние покупатели, которые не потратили много

Hold and improve - Раньше часто покупали, но давно не возвращались. Нужно вернуть их!

Risky - Последняя покупка была давно и было мало заказов, либо часто покупал, но давно. Может быть потерян.

Сгруппируем сегменты в кластеры и присвоим им названия


```python
def label_rfm_segments(RFM_Score):

    if (RFM_Score == '111') or (RFM_Score == '113') or (RFM_Score == '121'):
        return 'Risky'   

    elif (RFM_Score == '211') or (RFM_Score == '122') or (RFM_Score == '112'):
        return 'Hold and improve'       

    elif ((RFM_Score == '222')) or (RFM_Score == '213') or (RFM_Score == '212'):
        return 'Potential loyal'      

    elif (RFM_Score == '311') or (RFM_Score == '232') or (RFM_Score == '221'):
        return 'Loyal'      

    else:
        return 'VIP'
```

```python
RFM['Segment'] = RFM.apply(lambda x: label_rfm_segments(x.RFM_Score), axis=1)

RFM.tail()
```

Посчитаем число клиентов в каждом сегменте и визуализируем данные 


```python

segments_counts = RFM['Segment'].value_counts().sort_values(ascending=True)

fig, ax = plt.subplots()

bars = ax.barh(range(len(segments_counts)),
              segments_counts,
              color='silver')
ax.set_frame_on(False)
ax.tick_params(left=False,
               bottom=False,
               labelbottom=False)
ax.set_yticks(range(len(segments_counts)))
ax.set_yticklabels(segments_counts.index)

for i, bar in enumerate(bars):
        value = bar.get_width()
        if segments_counts.index[i] in ['champions', 'loyal customers']:
            bar.set_color('firebrick')
        ax.text(value,
                bar.get_y() + bar.get_height()/2,
                '{:,} ({:}%)'.format(int(value),
                                   int(value*100/segments_counts.sum())),
                va ='center',
                ha ='left'
               )

plt.show()
```
